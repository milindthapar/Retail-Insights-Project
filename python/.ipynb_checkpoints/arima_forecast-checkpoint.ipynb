```python
# ---------- 1. Imports ----------
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Make plots a bit nicer
plt.style.use("default")

# ---------- 2. Load cleaned data ----------
file_path = r"C:\Users\thapa\OneDrive\Desktop\Retail-Insights_Project\data\superstore_clean.csv"
df = pd.read_csv(file_path)

# Ensure date column is datetime
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

# ---------- 3. Build monthly revenue time series ----------
monthly = (
    df
    .set_index("Order_Date")["Sales"]
    .resample("M")          # monthly frequency
    .sum()
    .to_frame(name="MonthlyRevenue")
)

print("Monthly series shape:", monthly.shape)
display(monthly.head())

# ---------- 4. Fit a simple ARIMA model ----------
# You can tweak (p,d,q) if you want later; (1,1,1) is a solid starting point.
series = monthly["MonthlyRevenue"]

model = ARIMA(series, order=(1, 1, 1))
results = model.fit()

print(results.summary())

# ---------- 5. Forecast next 6 months ----------
steps = 6
forecast_obj = results.get_forecast(steps=steps)

forecast_mean = forecast_obj.predicted_mean
conf_int = forecast_obj.conf_int()   # lower / upper bounds

display(forecast_mean)
display(conf_int)

# ---------- 6. Plot historical + ARIMA forecast ----------
plt.figure(figsize=(12, 6))

# Historical
plt.plot(
    monthly.index,
    monthly["MonthlyRevenue"],
    label="Historical Revenue",
)

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean,
    label="ARIMA Forecast (next 6M)",
    linestyle="--",
)

# Confidence interval band
plt.fill_between(
    forecast_mean.index,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.2,
    label="95% Confidence Interval",
)

plt.title("Monthly Revenue Forecast  ARIMA(1,1,1)")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.tight_layout()

# Save figure for your report / portfolio
output_path = r"C:\Users\thapa\OneDrive\Desktop\Retail-Insights_Project\outputs\arima_forecast.png"
plt.savefig(output_path, dpi=300)
plt.show()

print("Figure saved to:", output_path)
```
